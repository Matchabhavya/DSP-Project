```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
iris=datasets.load_iris()
df=pd.DataFrame(data=iris.data,columns=iris.feature_names)
df['target']=iris.target
df
```

```python
df.sample(5)
```

```python
x=df.drop('target',axis=1)
y=df.target
def mink_dis(a,b,p=1):
    dim=len(a)
    dis=0
    for d in range(dim):
        dis+=abs(a[d]-b[d])**p
        dis =dis**(1/p)
        return dis
    mink_dis(a=x.iloc[0],b=x.iloc[1],p=1)
```

```python
test_pt=[4,6,3,4,3.5]
dist=[]
for i in x.index:
    dist.append(mink_dis(test_pt,x.iloc[i]))
df_dist=pd.DataFrame(data=dist,index=x.index,columns=['dist'])
df_dist.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=15)
clf=KNeighborsClassifier(n_neighbors=5,p=1)
clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
print(f'sklearn accuracy:{accuracy_score(y_test,y_pred)}')
```
